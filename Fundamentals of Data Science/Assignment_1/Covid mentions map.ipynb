```
import pandas as pd
import numpy as np
import geopandas as gpd
from shapely.geometry import Point
import matplotlib.pyplot as plt 
# import folium
# import plotly_express as px
```

```
world = gpd.read_file(gpd.datasets.get_path("naturalearth_lowres"))
world.head()
```

```
from google.colab import files
uploaded = files.upload()
```

```
import io
df = pd.read_csv(io.BytesIO(uploaded['out.csv']))
```

```
display(df)
```

```
world_happiness_final = world.merge(df, how="left", left_on=['iso_a3'], right_on=['ISO-alpha3 Code'])
world_happiness_final.head(10)
```

```
import matplotlib.pyplot as plt

world_happiness_final.plot("covid", 
                           figsize=(15,10),
                           legend=True,
                           edgecolor="white",
                           legend_kwds={"label":"covid mentions", "orientation":"horizontal"})
```
